### Working with Titianic Dataset

```python
# Import 

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
import pandas as pd
```

```python
# Create Our X and Y datasets 

titanic_x = pd.read_csv('titanic_train.csv')
titanic_y = titanic_x.pop('Survived')
```

```python
# Lets see some basic stats on data 

titanic_x.describe()
```

```python
# Let us impute age with the mean for the missing values. 

titanic_x['Age'].fillna(titanic_x.Age.mean(), inplace = True)


# Check out the stats again. 

titanic_x.describe()
```

```python
# Looking at the numercial variables

numeric_variables = list(titanic_x.dtypes[titanic_x.dtypes != 'object'].index)
titanic_x[numeric_variables].head()
```

```python
# Create our model here. 

model = RandomForestRegressor(n_estimators=100, oob_score=True, random_state=1)

# Fit the model.

model.fit(titanic_x[numeric_variables], titanic_y)
```

```python
# Let us examine the oob score. 

model.oob_score_
```
